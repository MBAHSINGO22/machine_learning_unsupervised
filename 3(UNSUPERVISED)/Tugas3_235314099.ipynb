```python
import pandas as pd
import numpy as np
import sklearn.metrics as sm
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
x=pd.DataFrame(df.data)
```

```python
x.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_width']
```

```python
x
```

```python
y = pd.DataFrame(df.target)
```

```python
y.columns=["Target"]
```

```python
y
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(x)
```

```python
model.labels_
```

```python
plt.scatter(x.Petal_Length, x.Petal_width)
```

```python
map = np.array(["Red","Green","Blue"])
```

```python
plt.scatter(x.Petal_Length, x.Petal_width,c=map[y.Target],s=40)
plt.title('Data Asli')
```

```python
plt.scatter(x.Petal_Length, x.Petal_width,c=map[model.labels_],s=40)
plt.title('K-means ')
```
